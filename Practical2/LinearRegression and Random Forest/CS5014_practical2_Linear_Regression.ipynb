```python
# Import packages for this practical
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import KFold
from sklearn import metrics
import sklearn.preprocessing as prep
```

```python
# Read data from file
test_data = pd.read_csv('data_test.csv')
train_data = pd.read_csv('data_train.csv')
```

```python
# Drop unwanted (Useless) labels
test_data = test_data.drop(axis=1, labels=['image', 'id'])
train_data = train_data.drop(axis=1, labels=['image', 'id'])
```

```python
# Encode Labels
df = pd.get_dummies(train_data['color'])
color_labels = []
for i in df:
    color_labels.append(i)
train_data = pd.concat([train_data, df], axis=1)

df = pd.get_dummies(train_data['texture'])
texture_labels = []
for i in df:
    texture_labels.append(i)
train_data = pd.concat([train_data, df], axis=1)

train_data = train_data.drop(axis=1, labels=['texture'])
train_data = train_data.drop(axis=1, labels=['color'])
```

```python
# Fix errors: drop duplicates, fill Nan for both train set and test set
train_data = train_data.drop_duplicates().dropna(axis=0, how='any').reset_index(drop=True)
test_data = test_data.fillna(value=0)
```

```python
# Normalize for training set
scale_parameter = {}
input_vars = ['w', 'h', 'x', 'y']
for i in train_data:
    for j in ['lightness_', 'redgreen_', 'blueyellow_', 'hog_', 'bimp_']:
        if j in i:
            input_vars.append(i)
            scale_parameter[i] = [train_data.loc[:, i].mean(), train_data.loc[:, i].std()]
            train_data.loc[:, i] = prep.scale(train_data[i])
```

```python
for i in test_data:
    if i in scale_parameter:
        test_data.loc[:, i] = test_data[i].map(lambda x: (x - scale_parameter[i][0]) / scale_parameter[i][1])
```

```python
# Train the model
def train_model(t_d):
    color_model = LinearRegression()
    color_model.fit(t_d.loc[:, input_vars], t_d.loc[:, color_labels])
    texture_model = LinearRegression()
    texture_model.fit(t_d.loc[:, input_vars], t_d.loc[:, texture_labels])
    return color_model, texture_model
```

```python
# Reverse coded series to the list of objects
def pred_verse_one_hot(coded, labels):
    res = []
    for i in coded:
        res.append(labels[np.where(i == np.max(i))[0][0]])
#         print(np.where(i == np.max(i))[0][0])
    return res
```

```python
# Reverse coded series back to the list of objects
def orig_verse_one_hot(coded, labels):
    res = []
    for i, r in coded.iterrows():
        for j in labels:
            if r[j] == 1:
                res.append(j)
                break
    return res
```

```python
# Model validation and evaluate performance of the model
def validate_model(model, v_d, area):
    res = model.predict(v_d.loc[:, input_vars])
#     print(res)
    c_l = color_labels if area == 'color' else texture_labels
    res_o = pred_verse_one_hot(res, c_l)
#     print(v_d.loc[:, c_l])
    rea_o = orig_verse_one_hot(v_d.loc[:, c_l], c_l)
    return metrics.accuracy_score(res_o, rea_o), metrics.balanced_accuracy_score(res_o, rea_o)
```

```python
# For training set, split for training set and validation set
color_accuracy, color_balanced_accuracy, texture_accuracy, texture_balanced_accuracy = 0, 0, 0, 0
kf = KFold(n_splits=5, shuffle=True)
for train_index, valid_index in kf.split(train_data):
    color_model, texture_model = train_model(train_data.loc[train_index, :])
    a, b = validate_model(color_model, train_data.loc[valid_index, :], "color")
    color_accuracy += a
    color_balanced_accuracy += b
    print(f'Color model evaluation: Accuracy - {a}, Balanced Accuracy - {b}')
    a, b = validate_model(texture_model, train_data.loc[valid_index, :], "texture")
    texture_accuracy += a
    texture_balanced_accuracy += b
    print(f'Texture model evaluation: Accuracy - {a}, Balanced Accuracy - {b}')
print(f'\n\nOverall evaluation: \nColor Accuracy: {color_accuracy / 5}, Color balanced accuracy: {color_balanced_accuracy / 5}')
print(f'Texture Accuracy: {texture_accuracy / 5}, Texture balanced accuracy: {texture_balanced_accuracy / 5}')
```

```python
# Get the final result
color_model, texture_model = train_model(train_data)
res = color_model.predict(test_data.loc[:, input_vars])
res_obj = pred_verse_one_hot(res, color_labels)
pd.DataFrame(res_obj).to_csv("color.csv", index=False, header=False)
res = texture_model.predict(test_data.loc[:, input_vars])
res_obj = pred_verse_one_hot(res, texture_labels)
pd.DataFrame(res_obj).to_csv("texture.csv", index=False, header=False)
```
